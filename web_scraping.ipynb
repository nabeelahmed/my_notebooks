# urllib2 
simple interface for network resource access. 
Purpose: Accessing remote resources that don’t need authentication, cookies, etc.

urlopen() - get a file-like handle to the remote resource.
gives access to the headers from the HTTP server through the info() method, and the data for the remote resource via methods like read() and readlines().

```python
import urllib2

response = urllib2.urlopen('http://google.com')
```

```python
print 'Response: ', response
print 'Url: ', response.geturl()
print '------------ Header Start -------------------'

headers = response.info()

#print 'Date:', headers['date']
print headers
print '------------ Header End --------------------------\n'


print '------------ Data Start -------------------'
data = response.read()
print data
print '------------ Data End --------------------------\n'
```

# Beautifulsoup
Python library for pulling data out of HTML and XML files.
Making a soup:

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("google_homepage.html"))
# soup = BeautifulSoup(open("google_homepage.html"), 'html.parser')
# soup = BeautifulSoup(open("google_homepage.html"), 'lxml')  # very fast
# soup = BeautifulSoup(open("google_homepage.html"), ['lxml', 'xml'])
# soup = BeautifulSoup(open("google_homepage.html"), 'html5lib')  # parses pages same as web-browser does. Creates valid html5
```

After openning the document, parses the document using the best available parser i.e. 
It will use an HTML parser unless you specifically tell it to use an XML parser.

```python
print soup
```

```python
print(soup.prettify())
```

```python
print 'Title - ', soup.title
print 'Title tag name - ', soup.title.name
print 'Title string- ', soup.title.string
print "Title tag's parent - ", soup.title.parent.name
```

```python
soup.p  # the first occurence of <p>
soup.find_all('p')  # all the paragraphs i.e. <p> tags in the document.
```

```python
soup.find_all('a')  # all the links i.e. <a> tags in the document.
```

```python
# pretty printing the links
for link in soup.find_all('a'):
    print link.get('href')
```

```python
# extracting all the text from the doc
# ISSUE: also gets the CSS and script statements !!!
print soup.get_text()
```

```python
# SOULTION - strip style and script elements
# https://stackoverflow.com/questions/22799990/beatifulsoup4-get-text-still-has-javascript
soup(['a', 'p'])
for style in soup(['style', 'script']):
    style.decompose()
    # script.extract()    # rip it out
```

```python
# verify - if script and style elements have been removed
print soup.get_text()
```

Can use Regex with BS4. For extracting non-standard data (not elements)

```python
import re


# all the tags starting with letter 'b'
for tag in soup.find_all(re.compile("^b")):
    print tag.name
    
```

```python
# all the tags having letter 'a'
for tag in soup.find_all(re.compile("a")):
    print tag.name
    
```
